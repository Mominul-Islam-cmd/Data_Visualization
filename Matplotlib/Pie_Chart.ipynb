```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
slice=[30,40,50,60,70]
L=['Thirty','Fourty','Fifity','SIxty','Seventy']
```

```python
plt.pie(slice,labels=L)
plt.title('Pie Chart')
plt.legend(['Thirty','Fourty','Fifity','SIxty','Seventy'])
```

```python
s=['pytohn','Machine Learning','Artiicial Intelligence','Deep learning','Natural processing Language','Matplotlib','Numpy','Pandas','Data science']
l=[123,311,578,434,605,700,800,850,1000]
plt.pie(l,labels=s,autopct='%.2f%%')
plt.legend(loc="upper left",bbox_to_anchor=(1.95,1.1))
```
