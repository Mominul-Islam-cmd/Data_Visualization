```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np 
import random
```

```python
Cse_students=np.random.randint(20,35,80)
Math_students=np.random.randint(17,40,80)
Cse_students
```

```python
Math_students
```

```python
plt.figure(figsize=(13,10))
plt.hist([Cse_students,Math_students],rwidth=0.9,orientation='vertical')
plt.title("Cse_students Age History_histogram",color='r',fontsize=20,weight='bold')
plt.xlabel('Age',color='r',fontsize=20,weight='bold')
plt.ylabel('No of Students',color='r',fontsize=20,weight='bold')
plt.legend(['Cse_students','Math_students'])
plt.show()
```

```python
plt.figure(figsize=(13,10))
style.use('ggplot')
plt.hist([Cse_students,Math_students],rwidth=0.9,orientation='vertical',edgecolor='black')
plt.title("Cse_students Age History_histogram",color='r',fontsize=20,weight=20)
plt.xlabel('Age',color='r',fontsize=20,weight='bold')
plt.ylabel('No of Students',color='r',fontsize=20,weight='bold')
plt.legend(['Cse_students','Math_students'])
plt.show()
```
