```python
Bin=[0.2,1.2,2,3.2,4.2,5.2,6.5,7.5,8.5,9.2,10.2,11.2,12.2]
```

```python
sns.set()
plt.figure(figsize=(10,7))
sns.distplot(df["tip"],bins=Bin,label="Tip",color="blue")
plt.xticks(Bin)
plt.legend()
plt.title("Tip",fontsize=20,color="r")
plt.show()
```

```python
df.total_bill.sort_values()
```

```python
Bins=[1,5,10,15,25,30,35,40,45,50,55,60]
```

```python
sns.set()
plt.figure(figsize=(10,7))
sns.distplot(df["total_bill"],bins=Bins,label="Total Bill",color="r")
plt.xticks(Bins)
plt.legend()
plt.title("Total Bill",fontsize=20,color="r")
plt.show()
```

```python
sns.set()
plt.figure(figsize=(10,7))
sns.distplot(df["total_bill"],bins=Bins,label="Total Bill",kde_kws={"color": "orange", "lw": 3, "label": "KDE"},  hist_kws={ "linewidth": 6,"alpha": 1, "color": "b","edgecolor":"black","linestyle":'--'})
plt.xticks(Bins)
plt.legend()
plt.title("Total Bill",fontsize=20,color="r")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.distplot(df["size"], label="size",color="green",bins=1)
sns.distplot(df["tip"],bins=1,label="Tip",color="blue")
sns.distplot(df["total_bill"],bins=1,label="Total Bill",color="r")
plt.xticks(Bins)
plt.xticks(Bin)
plt.show()
```
