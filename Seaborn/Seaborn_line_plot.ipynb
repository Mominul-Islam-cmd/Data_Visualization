```python
data1=sns.load_dataset('tips')
data1
```

```python
sns.lineplot('tip','total_bill',data=data1,hue='sex',style='sex')
```

```python
sns.lineplot('total_bill','tip',data=data1,hue='smoker',style='sex')
```

```python
plt.figure(figsize=(15,10))
sns.lineplot('total_bill','tip',data=data1,hue='smoker',style='sex',markers=["o",">"])
plt.xlabel("Total_bill",fontsize=15)
plt.ylabel("Tip",fontsize=15)
plt.title("Line Plot Tips data",fontsize=25)
```

```python
plt.figure(figsize=(15,10))
sns.lineplot('size','total_bill',data=data1,hue='day',style='day',markers=["o",">","<","*"])
plt.ylabel("Total_bill",fontsize=15)
plt.xlabel("Size",fontsize=15)
plt.title("Line Plot Tips data",fontsize=25)
```
